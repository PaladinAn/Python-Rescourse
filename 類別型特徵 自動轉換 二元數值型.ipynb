```python
%matplotlib inline

import numpy as np
import pandas as pd

from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.DataFrame([['A','windy','hot',25],['B','sunny','hot',30],['B','cloudy','cold',18]],columns=['Area','weather','feeling','temperature'])
df=data.copy()
```

```python
df
```

# OneHotEncoder   (PY版本不同，會有參數設定問題)

```python
enc = OneHotEncoder()
```

```python
df_one = enc.fit_transform(df)
```

```python
le = LabelEncoder()
for col in df[['Area','weather','feeling']]:
    df[col]=le.fit_transform(df[col])
df
```

```python
df_ohe = enc.fit_transform(df).toarray()      #因為 OneHotEncode，資料型態不同，要轉array
pd.DataFrame(df_ohe)                          #再轉DF
```

```python
enc_1 = OneHotEncoder()
```

```python
df_ohe=enc_1.fit_transform(df).toarray()
pd.DataFrame(df_ohe)
```

# get_dummies   自動做2元的拆分

```python
df = data.copy()
df
```

```python
df_dum = pd.get_dummies(df)
df_dum
```

```python
df_dum=pd.get_dummies(df[['Area','weather']])
df_dum
```

```python
df_new=pd.concat([df_dum,df['temperature']],axis=1)     #合併 溫度
df_new
```
