```python
X = np.random.rand(100,2)   #生100資料2個數字
```

```python
X[0]
```

```python
plt.scatter(X[:,0], X[:,1],s=50)
```

```python
from sklearn.cluster import KMeans
```

```python
clf = KMeans(n_clusters=3)
```

```python
clf.fit(X)             #學習
```

```python
clf.labels_
```
