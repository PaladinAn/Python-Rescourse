請安裝，pandas-datareader套件

    終端機執行，conda install pandas-datareader

    更新，conda update pandas-datareader

載入 pandas_datareader.data

```python
import pandas_datareader.data as web
```

```python
df = web.DataReader("AAPL", "yahoo", start="2012-9-1", end="2017-8-31")
```

取得蘋果，雅虎資料，日期

```python
df.head()
```

只要調整收盤價P = df["Adj Close"]

```python
P = df["Adj Close"]
```

```python
P.head()
```

```python
P.plot()
```

計算報酬率.diff()

```python
P.diff()
```

P.diff()/P，改百分比

```python
P.diff()/P
```

定義報酬率r=P.diff()/P

```python
r = P.diff()/P
```

```python
r.plot()
```

100天~現在：r[-100:].plot()

```python
r[-100:].plot()
```

移動平均：P.rolling(window=20).mean()，做20天

```python
P.rolling(window=20).mean()
```

```python
P.rolling(window=20).mean().plot()
```

要做比較，直接畫出來就好了

```python
P.plot()
P.rolling(window=20).mean().plot()
```

```python
P.plot()
P.rolling(window=20).mean().plot()
P.rolling(window=60).mean().plot()
```
