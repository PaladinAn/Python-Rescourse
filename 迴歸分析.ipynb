```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

準備模擬的資料

$$f(x)=1.2x+0.8$$

50個樣本

```python
x = np.linspace(0, 5, 50)     ###0~5，50個點###
```

繪圖

```python
y = 1.2*x + 0.8
```

```python
plt.scatter(x,y)     ####scatter點圖###
```

```python
plt.scatter(x,y)
plt.plot(x, y, "r")       ###圓滑曲線###
```

加入 noise 項，更接近現實

$$𝑓(𝑥)+ε(x)$$

```python
y = 1.2*x + 0.8 + 0.6*np.random.randn(50)
```

random   亂數隨機

randn  標準差0，平均數1

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x, 1.2*x + 0.8, "r")    #y已經加入，Noise，不能再代入y#
```

X原本是

$$[x1,x2,...,x50]$$

使用scikit-learn套件，格式要改成

$$[[x1],[x2],...,[x50]]$$

```python
from sklearn.linear_model import LinearRegression   ###載入scikit-learn套件###
```

先創一個，線性迴歸機

機器變數名稱 = LinearRegression()

```python
regr = LinearRegression()
```

```python
X = x.reshape(50,1)
```

變更array形狀，變數名稱.reshape(新的形狀)

```python
X
```

學習機的名稱.fit指令(input資料,正確資料)

```python
regr.fit(X, y)
```

```python
Y = regr.predict(X)
```

開始預測

機器名稱.predict(輸入的資料)

```python
plt.scatter(x, y)               ###散播圖，小y，正確答案###
plt.plot(x, Y, "r")             ###圓滑曲線，大Y，機器預測###
plt.plot(x, 1.2*x + 0.8, "g")   ###完美答案###
```
